```python
df = df.dropna(subset=['Income']).copy()
df['Income'] = df['Income'].fillna(df['Income'].median())
```

```python
df.isnull().sum() #null values removed in Income 
```

```python
df.duplicated().sum() 
```

```python
# There is no duplicated record found in this dataset so there is no need to perform drop_duplicates
```

```python
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()  
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({
    'married': 'married',
    'together': 'married',
    'single': 'single',
    'divorced': 'divorced',
    'widow': 'widow'
})
```

```python
df['Education'] = df['Education'].str.strip().str.upper()
```

```python
df
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df['age'] = 2024 - df['year_birth']

numeric_cols = ['kidhome', 'teenhome', 'recency', 'mntwines', 'mntfruits', 'mntmeatproducts',
                'mntfishproducts', 'mntsweetproducts', 'mntgoldprods', 'numdealspurchases',
                'numwebpurchases', 'numcatalogpurchases', 'numstorepurchases', 'numwebvisitsmonth']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)

```

```python
df = df.drop(columns=['income($)']) # deleted another income column created during the cleaning 
```
